```python
import pandas as pd
import random

# Load quiz data
quiz_data = pd.read_csv('Module1_Notes_to_Quiz/quiz.csv')

# Load enrollment numbers
enrollment_data = pd.read_csv('Test_og - Test_og.csv.csv')

# Initialize student data
num_students = 30
student_data = []

# Generate fake answers for each student using enrollment numbers
for student_id in range(num_students):
    # Get enrollment number
    enrollment_no = enrollment_data.iloc[student_id][0]
    student_answers = []
    
    for index, row in quiz_data.iterrows():
        # Randomly choose an answer from options A, B, C, D
        answer = random.choice(['A', 'B', 'C', 'D'])
        # Get the corresponding option text
        option_text = row[f'Option {answer}']
        # Format answer as "B) option"
        formatted_answer = f"{answer}) {option_text}"
        student_answers.append(formatted_answer)
    
    # Append enrollment number and their answers
    student_data.append([enrollment_no] + student_answers)

# Create a DataFrame for the new CSV
columns = ['Enrollment No'] + list(quiz_data['Question'])
new_quiz_df = pd.DataFrame(student_data, columns=columns)

# Save to new CSV file
new_quiz_df.to_csv('fake_student_quiz_data.csv', index=False)
```

```python
import pandas as pd

# Load the CSV files
quiz_df = pd.read_csv('Module1_Notes_to_Quiz/quiz.csv')
response_df = pd.read_csv('Module2_Quiz_Analysis/Test_Responce.csv')

# Initialize a dictionary to hold incorrect answers for each student
incorrect_answers = {}

# Iterate through each student's responses
for index, row in response_df.iterrows():
    enrollment_no = row['Enrollment No']
    incorrect_answers[enrollment_no] = []

    # Compare each response with the correct answer
    for i, question in enumerate(quiz_df['Question']):
        correct_answer = quiz_df['Correct Answer'][i]
        student_response = row[i + 1]  # Responses start from the second column

        if student_response != correct_answer:
            incorrect_answers[enrollment_no].append(question)

# Display the results
for enrollment_no, questions in incorrect_answers.items():
    # print(f"Enrollment No: {enrollment_no}, Incorrect Answers: {questions}")
    incorrect_ans = f"Enrollment No: {enrollment_no}, Incorrect Answers: {questions}"
    print(incorrect_ans)
```
